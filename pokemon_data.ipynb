```python
df.groupby = 'type 1'
```

```python
#Find Top 10 Pokemon with Highest Attack

top_attack = df[['Name', 'Attack']].sort_values(by='Attack', ascending=False).head(10)
print(top_attack)
```

```python
#Count Pokemon by Primary Type (Type 1)

type_counts = df['Type 1'].value_counts()
print(type_counts)

# Visualization
plt.figure(figsize=(10,6))
sns.barplot(x=type_counts.index, y=type_counts.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Number of PokÃ©mon by Primary Type')
plt.show()
```

```python
#Relationship Between Attack and Defense

plt.figure(figsize=(8,6))
sns.scatterplot(x = 'Attack', y = 'Defense', hue = 'Type 1', data = df)
plt.title('Attack vs Defense by Type')
plt.show()

```

```python
#Correlation Heatmap of Stats

plt.figure(figsize=(8,6))
sns.heatmap(df[['HP', 'Attack', 'Defense', 'Speed']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Stats')
plt.show()
```
